```python
import pickle
import numpy as np
import matplotlib.pyplot as plt

# ====================================
# LOAD DATA
# ====================================

with open("../Data/raw/WESAD/S2/S2.pkl", "rb") as f:
    data = pickle.load(f, encoding="latin1")

# ====================================
# EDA SIGNAL
# ====================================

signal = data['signal']['wrist']['EDA'].flatten()
labels = data['label']

print("EDA length:", len(signal))
print("Label length:", len(labels))
print("Unique labels:", np.unique(labels))

# ====================================
# MATCH LABELS TO EDA LENGTH
# ====================================

factor = len(labels) / len(signal)

indices = (np.arange(len(signal)) * factor).astype(int)
labels_resampled = labels[indices]

print("Resampled labels length:", len(labels_resampled))

# ====================================
# EXTRACT STATES
# ====================================

baseline = signal[labels_resampled == 1]
stress = signal[labels_resampled == 2]
meditation = signal[labels_resampled == 4]

print("\nSamples per state")
print("Baseline:", len(baseline))
print("Stress:", len(stress))
print("Meditation:", len(meditation))

# ====================================
# STATISTICS
# ====================================

states = ["Baseline", "Stress", "Meditation"]

means = [
    np.mean(baseline),
    np.mean(stress),
    np.mean(meditation)
]

stds = [
    np.std(baseline),
    np.std(stress),
    np.std(meditation)
]

print("\nStatistics")
for state, mean, std in zip(states, means, stds):
    print(f"{state}: Mean={mean:.4f}, Std={std:.4f}")

# ====================================
# BOXPLOT
# ====================================

plt.figure(figsize=(10,6))

plt.boxplot(
    [baseline, stress, meditation],
    tick_labels=states
)

plt.title("EDA Distribution by State")
plt.xlabel("State")
plt.ylabel("EDA")
plt.grid(True)

plt.show()

# ====================================
# BAR CHART
# ====================================

plt.figure(figsize=(10,6))

plt.bar(
    states,
    means,
    yerr=stds,
    capsize=8
)

plt.title("Mean EDA by State")
plt.xlabel("State")
plt.ylabel("Mean EDA")
plt.grid(True)

plt.show()

# ====================================
# OVERLAY PLOT
# ====================================

N = min(
    3000,
    len(baseline),
    len(stress),
    len(meditation)
)

plt.figure(figsize=(14,6))

plt.plot(
    baseline[:N],
    label="Baseline"
)

plt.plot(
    stress[:N],
    label="Stress"
)

plt.plot(
    meditation[:N],
    label="Meditation"
)

plt.title("EDA Comparison")
plt.xlabel("Sample Index")
plt.ylabel("EDA")
plt.legend()
plt.grid(True)

plt.show()

# ====================================
# DIFFERENCE FROM BASELINE
# ====================================

stress_diff = np.mean(stress) - np.mean(baseline)
med_diff = np.mean(meditation) - np.mean(baseline)

print("\nDifference from Baseline")
print(f"Stress Difference: {stress_diff:.4f}")
print(f"Meditation Difference: {med_diff:.4f}")
```
