```python
import pandas as pd
import numpy as np
from windowing_filtered import create_windows_filtered

# Empatica
eda_e4   = pd.read_csv('eda_filtered.csv')['EDA'].values
bvp_e4   = pd.read_csv('bvp_filtered.csv')['BVP'].values
hr_e4    = pd.read_csv('hr_filtered.csv')['HR'].values
temp_e4  = pd.read_csv('temp_filtered.csv')['TEMP'].values

# RespiBAN
ecg_resp  = pd.read_csv('ecg_filtered_respiban.csv')['ECG'].values
eda_resp  = pd.read_csv('eda_filtered_respiban.csv')['EDA'].values
emg_resp  = pd.read_csv('emg_centered_respiban.csv')['EMG'].values
resp_resp = pd.read_csv('resp_filtered_respiban.csv')['RESP'].values
temp_resp = pd.read_csv('temp_filtered_respiban.csv')['TEMP'].values

print(len(eda_e4), len(bvp_e4), len(hr_e4), len(temp_e4))
print(len(ecg_resp), len(eda_resp), len(emg_resp), len(resp_resp), len(temp_resp))
```

```python
result = create_windows_filtered(signals, fs_dict, window_sec=5, overlap=0.5)

for name, arr in result['windows'].items():
    print(name, arr.shape)
```

```python
ibi_df = pd.read_csv('ibi_filtered.csv')   # expects a time column + an IBI column
time_col, ibi_col = ibi_df.columns[0], ibi_df.columns[1]

window_sec, overlap = 5, 0.5
step_sec = window_sec * (1 - overlap)

ibi_windows = []
t_start = ibi_df[time_col].iloc[0]
t_end = ibi_df[time_col].iloc[-1]

t = t_start
while t + window_sec <= t_end:
    mask = (ibi_df[time_col] >= t) & (ibi_df[time_col] < t + window_sec)
    ibi_windows.append(ibi_df.loc[mask, ibi_col].values)
    t += step_sec

print(len(ibi_windows), 'IBI windows,', 'e.g. window 0 has', len(ibi_windows[0]), 'beats')
```
