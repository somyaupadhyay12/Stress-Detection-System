```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read the CSV file
df = pd.read_csv("all_model_comparisons.csv")   # Replace with your CSV file path

# Metrics to compare
metrics = [
    "test_f1",
    "test_precision",
    "test_recall",
    "test_accuracy"
]

# Create grouped bar chart
x = np.arange(len(df))
width = 0.2

plt.figure(figsize=(12, 6))

for i, metric in enumerate(metrics):
    plt.bar(x + i*width, df[metric], width,
            label=metric.replace("test_", "").replace("_", " ").title())

# Formatting
plt.xticks(
    x + width*1.5,
    df["model"].str.replace("_", " ").str.title(),
    rotation=15
)

plt.ylabel("Score")
plt.xlabel("Models")
plt.title("Comparison of Model Performance")
plt.ylim(0, 1.05)
plt.legend()
plt.grid(axis="y", linestyle="--", alpha=0.5)

plt.tight_layout()
plt.show()
```
