```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# E4 EDA
file_path = ".../data/raw/WESAD/S2/S2_E4_Data/EDA.csv"
e4_eda = pd.read_csv(file_path, header=None).values

# Respiban
resp = pd.read_csv(
    "data/raw/WESAD/S2/S2_respiban.txt",
    sep=r"\s+",
    comment="#",
    header=None
)

# Respiban EDA column
resp_eda = resp.iloc[:, 3].values
time_e4 = np.arange(len(e4_eda)) / 4      # E4 EDA = 4 Hz
time_resp = np.arange(len(resp_eda)) / 700 # Respiban = 700 Hz
plt.figure(figsize=(12,5))

plt.plot(time_e4, e4_eda[0], label="E4 EDA")
plt.plot(time_resp, resp_eda, label="Respiban EDA")

plt.xlabel("Time (seconds)")
plt.ylabel("EDA")
plt.title("E4 vs Respiban EDA")
plt.legend()
plt.show()
```

```python
 ##E4 VS RESPIBAN EDA 
import pickle
import numpy as np
import matplotlib.pyplot as plt

# Load subject data
with open("../data/raw/WESAD/S2/S2.pkl", "rb") as f:
    data = pickle.load(f, encoding="latin1")

# Extract signals
e4_eda = data["signal"]["wrist"]["EDA"]
resp_eda = data["signal"]["chest"]["EDA"]

# Convert to numpy arrays
e4_eda = np.array(e4_eda).flatten()
resp_eda = np.array(resp_eda).flatten()

# Normalize both signals to 0-1
e4_eda = (e4_eda - np.min(e4_eda)) / (np.max(e4_eda) - np.min(e4_eda))
resp_eda = (resp_eda - np.min(resp_eda)) / (np.max(resp_eda) - np.min(resp_eda))

# Create separate time axes
e4_time = np.arange(len(e4_eda)) / 4      # E4 EDA = 4 Hz
resp_time = np.arange(len(resp_eda)) / 700 # Respiban EDA = 700 Hz

# Plot
plt.figure(figsize=(12,5))
plt.plot(e4_time, e4_eda, label="E4 EDA")
plt.plot(resp_time, resp_eda, label="Respiban EDA", alpha=0.7)

plt.xlabel("Time (seconds)")
plt.ylabel("Normalized EDA")
plt.title("E4 vs Respiban EDA")
plt.legend()
plt.grid(True)

plt.show()
```

```python
import pickle
import numpy as np
import matplotlib.pyplot as plt

# Load data
with open("data/raw/WESAD/S2/S2.pkl", "rb") as f:
    data = pickle.load(f, encoding="latin1")

# ---------------------------
# Helper function
# ---------------------------
def normalize(signal):
    signal = np.array(signal).flatten()
    return (signal - signal.min()) / (signal.max() - signal.min())

# ---------------------------
# TEMPERATURE
# ---------------------------
e4_temp = normalize(data["signal"]["wrist"]["TEMP"])
chest_temp = normalize(data["signal"]["chest"]["Temp"])

plt.figure(figsize=(12,4))
plt.plot(e4_temp, label="E4 TEMP")
plt.plot(chest_temp, label="Respiban TEMP")
plt.title("Temperature Comparison")
plt.ylabel("Normalized Value")
plt.legend()
plt.grid(True)
plt.show() 

# ---------------------------
# ACCELERATION
# ---------------------------
e4_acc = data["signal"]["wrist"]["ACC"]
chest_acc = data["signal"]["chest"]["ACC"]

# Combine X,Y,Z into magnitude
e4_acc_mag = np.sqrt(
    e4_acc[:,0]**2 +
    e4_acc[:,1]**2 +
    e4_acc[:,2]**2
)

chest_acc_mag = np.sqrt(
    chest_acc[:,0]**2 +
    chest_acc[:,1]**2 +
    chest_acc[:,2]**2
)

e4_acc_mag = normalize(e4_acc_mag)
chest_acc_mag = normalize(chest_acc_mag)

plt.figure(figsize=(12,4))
plt.plot(e4_acc_mag, label="E4 ACC")
plt.plot(chest_acc_mag, label="Respiban ACC")
plt.title("Acceleration Comparison")
plt.ylabel("Normalized Value")
plt.legend()
plt.grid(True)
plt.show()

e4_eda = normalize(data["signal"]["wrist"]["EDA"])
chest_eda = normalize(data["signal"]["chest"]["EDA"])

plt.figure(figsize=(12,4))
plt.plot(e4_eda, label="E4 EDA")
plt.plot(chest_eda, label="Respiban EDA")
plt.title("EDA Comparison")
plt.ylabel("Normalized Value")
plt.legend()
plt.grid(True)
plt.show()
```
