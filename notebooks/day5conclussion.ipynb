```python
import pickle
import numpy as np
import matplotlib.pyplot as plt

# ====================================
# LOAD WESAD DATA
# ====================================

with open("../Data/raw/WESAD/S2/S2.pkl", "rb") as f:
    data = pickle.load(f, encoding="latin1")

# ====================================
# CHOOSE SENSOR
# Change EDA to TEMP, BVP, etc. if needed
# ====================================

signal = data['signal']['wrist']['EDA'].flatten()

# Original labels (700 Hz)
labels = data['label']

# ====================================
# BASIC INFO
# ====================================

print("Signal length:", len(signal))
print("Label length:", len(labels))
print("Unique labels:", np.unique(labels))

# ====================================
# RESAMPLE LABELS TO MATCH SIGNAL LENGTH
# ====================================

factor = len(labels) / len(signal)

indices = (np.arange(len(signal)) * factor).astype(int)
labels_resampled = labels[indices]

print("Resampled label length:", len(labels_resampled))

# ====================================
# EXTRACT STATES
# ====================================

baseline = signal[labels_resampled == 1]
stress = signal[labels_resampled == 2]
amusement = signal[labels_resampled == 3]
meditation = signal[labels_resampled == 4]

print("\nSamples per state")
print("Baseline:", len(baseline))
print("Stress:", len(stress))
print("Amusement:", len(amusement))
print("Meditation:", len(meditation))

# ====================================
# STATISTICS
# ====================================

states = [
    "Baseline",
    "Stress",
    "Amusement",
    "Meditation"
]

means = [
    np.mean(baseline),
    np.mean(stress),
    np.mean(amusement),
    np.mean(meditation)
]

stds = [
    np.std(baseline),
    np.std(stress),
    np.std(amusement),
    np.std(meditation)
]

print("\n===== STATISTICS =====")

for state, mean, std in zip(states, means, stds):
    print(
        f"{state}: "
        f"Mean={mean:.4f}, "
        f"Std={std:.4f}"
    )

# ====================================
# BOXPLOT
# ====================================

plt.figure(figsize=(12, 6))

plt.boxplot(
    [baseline, stress, amusement, meditation],
    tick_labels=states
)

plt.title("EDA Distribution by State")
plt.xlabel("State")
plt.ylabel("EDA")
plt.grid(True)

plt.show()

# ====================================
# BAR CHART (MEAN Â± STD)
# ====================================

plt.figure(figsize=(12, 6))

plt.bar(
    states,
    means,
    yerr=stds,
    capsize=8
)

plt.title("Mean EDA by State")
plt.xlabel("State")
plt.ylabel("EDA")
plt.grid(True)

plt.show()

# ====================================
# OVERLAY COMPARISON
# ====================================

N = min(
    3000,
    len(baseline),
    len(stress),
    len(amusement),
    len(meditation)
)

plt.figure(figsize=(14, 6))

plt.plot(
    baseline[:N],
    label="Baseline"
)

plt.plot(
    stress[:N],
    label="Stress"
)

plt.plot(
    amusement[:N],
    label="Amusement"
)

plt.plot(
    meditation[:N],
    label="Meditation"
)

plt.title("EDA Signal Comparison")
plt.xlabel("Sample Index")
plt.ylabel("EDA")
plt.legend()
plt.grid(True)

plt.show()

# ====================================
# DIFFERENCE FROM BASELINE
# ====================================

stress_diff = np.mean(stress) - np.mean(baseline)
amusement_diff = np.mean(amusement) - np.mean(baseline)
meditation_diff = np.mean(meditation) - np.mean(baseline)

print("\n===== DIFFERENCE FROM BASELINE =====")

print(f"Stress Difference: {stress_diff:.4f}")
print(f"Amusement Difference: {amusement_diff:.4f}")
print(f"Meditation Difference: {meditation_diff:.4f}")

# ====================================
# HISTOGRAMS
# ====================================

plt.figure(figsize=(12, 6))

plt.hist(
    baseline,
    bins=50,
    alpha=0.5,
    label="Baseline"
)

plt.hist(
    stress,
    bins=50,
    alpha=0.5,
    label="Stress"
)

plt.hist(
    amusement,
    bins=50,
    alpha=0.5,
    label="Amusement"
)

plt.hist(
    meditation,
    bins=50,
    alpha=0.5,
    label="Meditation"
)

plt.title("EDA Histogram by State")
plt.xlabel("EDA")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)

plt.show()
```

```python
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ====================================
# LOAD LABELS FROM S2.pkl
# ====================================

with open("../Data/raw/WESAD/S2/S2.pkl", "rb") as f:
    data = pickle.load(f, encoding="latin1")

labels = data['label']

# ====================================
# LOAD HR SIGNAL
# ====================================

hr_df = pd.read_csv(
    "../Data/raw/WESAD/S2/S2_E4_Data/HR.csv",
    header=None
)

signal = hr_df.values.flatten()

# ====================================
# CHECK DATA
# ====================================

print("HR Length:", len(signal))
print("Label Length:", len(labels))

# ====================================
# MATCH LABELS TO HR LENGTH
# ====================================

factor = len(labels) / len(signal)

indices = (np.arange(len(signal)) * factor).astype(int)

labels_resampled = labels[indices]

# ====================================
# EXTRACT STATES
# ====================================

baseline = signal[labels_resampled == 1]
stress = signal[labels_resampled == 2]
amusement = signal[labels_resampled == 3]
meditation = signal[labels_resampled == 4]

# ====================================
# PRINT STATISTICS
# ====================================

states = ["Baseline", "Stress", "Amusement", "Meditation"]

means = [
    np.mean(baseline),
    np.mean(stress),
    np.mean(amusement),
    np.mean(meditation)
]

print("\n===== MEAN HR =====")

for s, m in zip(states, means):
    print(f"{s}: {m:.2f}")

# ====================================
# GRAPH 1: MEAN HR BY STATE
# ====================================

plt.figure(figsize=(10,6))

plt.plot(
    states,
    means,
    marker='o',
    linewidth=3
)

plt.title("Mean Heart Rate Across Emotional States")
plt.xlabel("Emotional State")
plt.ylabel("Heart Rate (BPM)")
plt.grid(True)

plt.show()

# ====================================
# GRAPH 2: BOXPLOT
# ====================================

plt.figure(figsize=(10,6))

plt.boxplot(
    [baseline, stress, amusement, meditation],
    tick_labels=states
)

plt.title("Heart Rate Distribution by State")
plt.xlabel("State")
plt.ylabel("Heart Rate (BPM)")
plt.grid(True)

plt.show()

# ====================================
# GRAPH 3: HISTOGRAM
# ====================================

plt.figure(figsize=(12,6))

plt.hist(
    baseline,
    bins=30,
    alpha=0.5,
    label="Baseline"
)

plt.hist(
    stress,
    bins=30,
    alpha=0.5,
    label="Stress"
)

plt.hist(
    amusement,
    bins=30,
    alpha=0.5,
    label="Amusement"
)

plt.hist(
    meditation,
    bins=30,
    alpha=0.5,
    label="Meditation"
)

plt.title("Heart Rate Histogram by State")
plt.xlabel("Heart Rate (BPM)")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)

plt.show()
```

```python
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ====================================
# LOAD LABELS
# ====================================

with open("../Data/raw/WESAD/S2/S2.pkl", "rb") as f:
    data = pickle.load(f, encoding="latin1")

labels = data['label']

# ====================================
# LOAD IBI
# ====================================

ibi = pd.read_csv(
    "../Data/raw/WESAD/S2/S2_E4_Data/IBI.csv",
    header=None,
    skiprows=1      # Skip timestamp row
)

time = ibi.iloc[:, 0].values
ibi_values = ibi.iloc[:, 1].values

print("IBI Samples:", len(ibi_values))
print("Label Samples:", len(labels))

# ====================================
# MATCH LABELS TO IBI LENGTH
# ====================================

factor = len(labels) / len(ibi_values)

indices = (np.arange(len(ibi_values)) * factor).astype(int)

ibi_labels = labels[indices]

# ====================================
# EXTRACT STATES
# ====================================

baseline = ibi_values[ibi_labels == 1]
stress = ibi_values[ibi_labels == 2]
amusement = ibi_values[ibi_labels == 3]
meditation = ibi_values[ibi_labels == 4]

# ====================================
# MEAN VALUES
# ====================================

states = [
    "Baseline",
    "Stress",
    "Amusement",
    "Meditation"
]

means = [
    np.mean(baseline),
    np.mean(stress),
    np.mean(amusement),
    np.mean(meditation)
]

stds = [
    np.std(baseline),
    np.std(stress),
    np.std(amusement),
    np.std(meditation)
]

print("\n===== IBI STATISTICS =====")

for s, m, sd in zip(states, means, stds):
    print(f"{s}: Mean={m:.4f}, Std={sd:.4f}")

# ====================================
# GRAPH 1
# MEAN IBI BY STATE
# ====================================

plt.figure(figsize=(10,6))

plt.plot(
    states,
    means,
    marker='o',
    linewidth=3
)

plt.title("Mean IBI Across Emotional States")
plt.xlabel("Emotional State")
plt.ylabel("Mean IBI (seconds)")
plt.grid(True)

plt.show()

# ====================================
# GRAPH 2
# BOXPLOT
# ====================================

plt.figure(figsize=(10,6))

plt.boxplot(
    [baseline, stress, amusement, meditation],
    tick_labels=states
)

plt.title("IBI Distribution by State")
plt.xlabel("State")
plt.ylabel("IBI (seconds)")
plt.grid(True)

plt.show()

# ====================================
# GRAPH 3
# HISTOGRAM
# ====================================

plt.figure(figsize=(12,6))

plt.hist(baseline, bins=30, alpha=0.5, label="Baseline")
plt.hist(stress, bins=30, alpha=0.5, label="Stress")
plt.hist(amusement, bins=30, alpha=0.5, label="Amusement")
plt.hist(meditation, bins=30, alpha=0.5, label="Meditation")

plt.title("IBI Histogram by State")
plt.xlabel("IBI (seconds)")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)

plt.show()

# ====================================
# GRAPH 4
# STATE COMPARISON BAR CHART
# ====================================

plt.figure(figsize=(10,6))

plt.bar(
    states,
    means,
    yerr=stds,
    capsize=8
)

plt.title("Mean IBI by Emotional State")
plt.xlabel("State")
plt.ylabel("Mean IBI (seconds)")
plt.grid(True)

plt.show()
```

```python
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ====================================
# LOAD LABELS
# ====================================

with open("../Data/raw/WESAD/S2/S2.pkl", "rb") as f:
    data = pickle.load(f, encoding="latin1")

labels = data['label']

# ====================================
# LOAD HR
# ====================================

hr_df = pd.read_csv(
    "../Data/raw/WESAD/S2/S2_E4_Data/HR.csv",
    header=None
)

hr = hr_df.values.flatten()

# ====================================
# LOAD IBI
# ====================================

ibi_df = pd.read_csv(
    "../Data/raw/WESAD/S2/S2_E4_Data/IBI.csv",
    header=None,
    skiprows=1
)

ibi_time = ibi_df.iloc[:, 0].values
ibi = ibi_df.iloc[:, 1].values

# ====================================
# ALIGN LABELS TO HR
# ====================================

factor_hr = len(labels) / len(hr)

hr_indices = (
    np.arange(len(hr)) * factor_hr
).astype(int)

hr_labels = labels[hr_indices]

# ====================================
# ALIGN LABELS TO IBI
# ====================================

factor_ibi = len(labels) / len(ibi)

ibi_indices = (
    np.arange(len(ibi)) * factor_ibi
).astype(int)

ibi_labels = labels[ibi_indices]

# ====================================
# CREATE TIME AXIS
# ====================================

hr_time = np.arange(len(hr))

# ====================================
# PLOT
# ====================================

fig, ax1 = plt.subplots(figsize=(16,8))

# ---------------------------
# HR
# ---------------------------

ax1.plot(
    hr_time,
    hr,
    linewidth=2,
    label="HR"
)

ax1.set_xlabel("Sample Index")
ax1.set_ylabel("Heart Rate (BPM)")

# ---------------------------
# IBI
# ---------------------------

ax2 = ax1.twinx()

ax2.plot(
    np.linspace(0, len(hr), len(ibi)),
    ibi,
    linewidth=2,
    alpha=0.8,
    label="IBI"
)

ax2.set_ylabel("IBI (seconds)")

# ====================================
# LABEL REGIONS
# ====================================

current_label = hr_labels[0]
start = 0

label_colors = {
    1: "green",    # Baseline
    2: "red",      # Stress
    3: "orange",   # Amusement
    4: "blue"      # Meditation
}

label_names = {
    1: "Baseline",
    2: "Stress",
    3: "Amusement",
    4: "Meditation"
}

for i in range(1, len(hr_labels)):

    if hr_labels[i] != current_label:

        ax1.axvspan(
            start,
            i,
            alpha=0.15,
            color=label_colors.get(current_label, "gray")
        )

        start = i
        current_label = hr_labels[i]

# Last region

ax1.axvspan(
    start,
    len(hr_labels),
    alpha=0.15,
    color=label_colors.get(current_label, "gray")
)

# ====================================
# LEGEND
# ====================================

from matplotlib.patches import Patch

legend_items = [
    Patch(color="green", alpha=0.3, label="Baseline"),
    Patch(color="red", alpha=0.3, label="Stress"),
    Patch(color="orange", alpha=0.3, label="Amusement"),
    Patch(color="blue", alpha=0.3, label="Meditation")
]

ax1.legend(handles=legend_items, loc="upper left")

plt.title(
    "HR and IBI with Emotional State Labels"
)

plt.grid(True)

plt.show()
```
