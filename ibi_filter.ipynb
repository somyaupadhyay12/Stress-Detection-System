```python
import pandas as pd

ibi_df = pd.read_csv(
    "../data/raw/WESAD/S2/S2_E4_Data/IBI.csv",
    header=None
)

print(ibi_df.head())
print(ibi_df.shape)
```

```python
print(ibi_df.head(10))
```

```python
start_time = ibi_df.iloc[0,0]

ibi_times = ibi_df.iloc[1:,0].astype(float).values
ibi_values = ibi_df.iloc[1:,1].astype(float).values

print("Number of IBI samples:", len(ibi_values))
print("Mean IBI:", ibi_values.mean())
print("Min IBI:", ibi_values.min())
print("Max IBI:", ibi_values.max())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,4))

plt.plot(ibi_times, ibi_values)

plt.title("Inter-Beat Interval (IBI)")
plt.xlabel("Time (s)")
plt.ylabel("IBI (s)")
plt.grid(True)

plt.show()
```

```python
downsample_factor = 5

ibi_times_ds = ibi_times[::downsample_factor]
ibi_values_ds = ibi_values[::downsample_factor]

plt.figure(figsize=(12,4))
plt.plot(ibi_times_ds, ibi_values_ds)
plt.grid(True)
plt.show()
```

```python
N = 500

plt.figure(figsize=(12,4))
plt.plot(ibi_times[:N], ibi_values[:N])
plt.grid(True)
plt.show()
```
